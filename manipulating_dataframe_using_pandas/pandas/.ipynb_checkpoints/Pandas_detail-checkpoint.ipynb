#Filtering DataFrames

#Combining filters

```python
df[(df.mpg >= 30) & (df.am < 2)] # Both conditions
```

```python
df[(df.mpg >= 30) | (df.am < 2)] # Either condition
```

#Combining filters

```python
df[(df.am == 1) & (df.mpg > 25)] # Both conditions
```

```python
df[(df.cyl<2) | (df.mpg < 20)] # Either condition
```

#Transforming DataFrames

##DataFrame vectorized methods

```python
df.floordiv(12) # Convert to dozens unit
```

#NumPy vectorized functions

```python
import numpy as np
np.floor_divide(df, 12) # Convert to dozens unit
```

#Plain Python functions (1)

```python
def dozens(n):
    return n//12
df.apply(dozens) # Convert to dozens unit
```

#Plain Python functions (2)

```python
df.apply(lambda n: n//12)
```

#Storing a transformation

```python
df['dozens_of_mpg'] = df.mpg.floordiv(12)
```
